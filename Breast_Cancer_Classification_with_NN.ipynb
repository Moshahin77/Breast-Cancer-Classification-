**Import dependences**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

**Data collection & processing**

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
df = pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
```

```python
#adding the label column to the dataframe
df["label"] = breast_cancer_dataset.target
```

# Seprating the data into features and target

```python
x = df.drop(columns ="label" , axis= 1 )
y = df["label"]
```

```python
print("The features (first 5 rows):")
print(x.head())
```

```python
print("The label (first 5 rows):")
print(y.head())
```

# **Spliting data into test data and training data**

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
  print(x.shape,x_train.shape,x_test.shape)
  print(y.shape,y_train.shape,y_test.shape)
```

# **Standardize the data**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train_std = scaler.fit_transform(x_train)
x_test_std = scaler.fit_transform(x_test)
```

# ***Building the neural network***

download.png

```python
#importing tensor flow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#settimg up the layers of neural network
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (30,)),
    keras.layers.Dense(20,activation = "relu"),
    keras.layers.Dense(2,activation= "sigmoid")
])
```

```python
# compiling the Neural Network
model.compile(optimizer = "adam",
              loss ="sparse_categorical_crossentropy",
              metrics = ["accuracy"])
```

```python
#training the neural network
history = model.fit(x_train_std,y_train,validation_split=0.1,epochs=10)
```

# **Visualizing accuracy and loss**

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
plt.show()
```

# **Accuracy of the data on the test data**

```python
loss, accuracy = model.evaluate(x_test_std, y_test)
print(accuracy)
```

# **Building the predictive system**

```python
input_data = (11.76,21.6,74.72,427.9,0.08637,0.04966,0.01657,0.01115,0.1495,0.05888,0.4062,1.21,2.635,28.47,0.005857,0.009758,0.01168,0.007445,0.02406,0.001769,12.98,25.72,82.98,516.5,0.1085,0.08615,0.05523,0.03715,0.2433,0.06563)

# change the input_data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardizing the input data
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if(prediction_label[0] == 0):
  print('The tumor is Malignant')

else:
  print('The tumor is Benign')
```
